```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'} 
webpage=requests.get('https://www.ambitionbox.com/list-of-companies?page=1',headers=headers).text
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
company=soup.find_all('div',class_='companyCardWrapper__primaryInformation')
```

```python
len(company)
```

```python
name=[]
rating=[]
reviews=[]
ctype=[]
for i in company:
   name.append(i.find('h2').text.strip())
   rating.append(i.find('div',class_='rating_star_container').text.strip())
   reviews.append(i.find('span',class_='companyCardWrapper__companyRatingCount').text.strip())
   ctype.append(i.find_all('span',class_='companyCardWrapper__interLinking')[0].text.strip())

d={'name':name,'rating':rating,'reviews':reviews,'Type':ctype}
df=pd.DataFrame(d)
df
    
```

```python
final=pd.DataFrame()
for j in range(1,50):
    url='https://www.ambitionbox.com/list-of-companies?page={}'.format(j)
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'} 
    webpage=requests.get(url,headers=headers).text
    soup=BeautifulSoup(webpage,'lxml')
    company=soup.find_all('div',class_='companyCardWrapper__primaryInformation')

    name=[]
    rating=[]
    reviews=[]
    ctype=[]
    all_data = [] 
    for i in company:
        name.append(i.find('h2').text.strip())
        rating.append(i.find('div',class_='rating_star_container').text.strip())
        reviews.append(i.find('span',class_='companyCardWrapper__companyRatingCount').text.strip())
        ctype.append(i.find_all('span',class_='companyCardWrapper__interLinking')[0].text.strip())

    d={'name':name,'rating':rating,'reviews':reviews,'Type':ctype}
    df=pd.DataFrame(d)
    all_data.append(df)
    final = pd.concat(all_data, ignore_index=True)
    
    
```

```python
print(final.to_string())
```
